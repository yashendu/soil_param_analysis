```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns

import warnings, random

random.seed(666)
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
path = os.getcwd()
in_path = path + "\\input\\"
out_path = path + "\\output\\"

# Check the files available in input directory by running this code

for dirname, _, filenames in os.walk(in_path):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
jhum = pd.read_excel(in_path+"Jhum_consolidated.xlsx")
```

```python
jhum.head()
```

```python
jhum.info()
```

```python
# Find the pearson correlations matrix
corr = jhum.corr(method = 'pearson')
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
## Defining filter function to split Jhum dataset parameter wise
## Colnames have been pre-processed in consolidated excel itself as part of data prep
## Colunm

def filter_col(df, str):
    c_all = list(df)
    c_new = ['rf', 'yr', 'dt']
    for s in c_all:
        if s.endswith(str):
            c_new.append(s)
    return c_new
```

```python
str = '_f'
jhum_col_f = filter_col(jhum, str)
jhum_col_f 
```

```python
jhum_f = jhum[jhum_col_f]
jhum_f 
```

```python
temp = jhum_f.drop(columns=['yr', 'dt'])
# Find the pearson correlations matrix
corr = temp.corr(method = 'pearson')
#corr
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
